### Task 10: Create a figure of 6 subplots in 2 rows and 3 columns of following functions. (10 points)
1. $y=sin(x)$, x range from -10 to 10, use np.sin(x)
2. $y=cos(x)$, x range from -10 to 10, use np.cos(x)
3. $y=x^2$,  x range from -5 to 5, use x\**2
4. $y=e^x$, x range from -5 to 5,use np.exp(x)
5. $y=log(x)$, x range from 0.1 to 100, use np.log(x)
6. $y=\frac{e^x}{e^x+1}$, x range from -10 to 10, use np.exp(x)/(np.exp(x)+1)

Requirements:
1. use `subplot(nrows,ncols,number)`
1. set `figsize = (16,10)`
2. Set the title as the function, you can double click this cell to copy how you can display the functions in Latex format (e.g. `$y=sin(x)$`). Put a 'r' befor the string, e.g., you should use `title(r'$y=sin(x)$')`
3. set line of each subplot with different colors
4. use x and y for axis labels.


```python
x = linspace(-10, 10)
y = np.sin(x)

fig = figure(figsize=(16,10))
subplot(2,3,1)
plot(x, y, 'r--') #plot x and y with dashed line in red
xlabel('x')
ylabel('y')
title(r'$y=sin(x)$');

y = np.cos(x)
subplot(2,3,2)
plot(x, y, 'k') #plot x and y with dashed line in red
xlabel('x')
ylabel('y')
title(r'$y=cos(x)$');

x = linspace(-5, 5)
y = x**2
subplot(2,3,3)
plot(x, y, 'c') #plot x and y with dashed line in red
xlabel('x')
ylabel('y')
title(r'$y=x^2$');

y = np.exp(x)
subplot(2,3,4)
plot(x, y, 'y') #plot x and y with dashed line in red
xlabel('x')
ylabel('y')
title(r'$y=e^x$');

x = linspace(.1, 100)
y = np.log(x)
subplot(2,3,5)
plot(x, y, 'b') #plot x and y with dashed line in red
xlabel('x')
ylabel('y')
title(r'$y=log(x)$');

x = linspace(-10, 10)
y = np.exp(x)/(np.exp(x)+1)
subplot(2,3,6)
plot(x, y, 'g') #plot x and y with dashed line in red
xlabel('x')
ylabel('y')
title(r'$y=\frac{e^x}{e^x+1}$');
```

### Task 11: Redo Task 10 with `plt.subplots(nrows,ncols)` (10 points)
1. $y=sin(x)$, x range from -10 to 10, use np.sin(x)
2. $y=cos(x)$, x range from -10 to 10, use np.cos(x)
3. $y=x^2$,  x range from -5 to 5, use **np.square(x)**
4. $y=e^x$, x range from -5 to 5,use np.exp(x)
5. $y=log(x)$, x range from 0.1 to 100, use np.log(x)
6. $y=\frac{e^x}{e^x+1}$, x range from -10 to 10, use np.exp(x)/(np.exp(x)+1)

**More Requirements**:
1. use `fig,axes = plt.subplots(nrows,ncols)`
1. set `figsize = (16,10)`
2. Set the title as the function, you can double click this cell to copy how you can display the functions in Latex format (e.g. `$y=sin(x)$`). Put a 'r' befor the string, e.g., you should use `title(r'$y=sin(x)$')`
3. set line of each subplot with different colors
3. set line of each subplot with different line types.
4. use x and y for axis labels.
5. use `fig.tight_layout()`



```python
fig, [[a1,a2,a3],[a4,a5,a6]] = plt.subplots(2, 3,figsize=(16,10))

x = linspace(-10, 10)
a1.plot(x, np.sin(x), 'r')
a1.set_xlabel('x')
a1.set_ylabel('y')
a1.set_title(r'$y=sin(x)$');

a2.plot(x, np.cos(x), 'y')
a2.set_xlabel('x')
a2.set_ylabel('y')
a2.set_title(r'$y=cos(x)$');

x = linspace(-5, 5)
a3.plot(x, x**2, 'g')
a3.set_xlabel('x')
a3.set_ylabel('y')
a3.set_title(r'$y=x^2$');

a4.plot(x, np.exp(x), 'b')
a4.set_xlabel('x')
a4.set_ylabel('y')
a4.set_title('$y=e^x$');

x = linspace(.1, 100)
a5.plot(x, np.log(x), 'c')
a5.set_xlabel('x')
a5.set_ylabel('y')
a5.set_title(r'$y=log(x)$');

a6.plot(x, np.exp(x)/(np.exp(x)+1), 'k')
a6.set_xlabel('x')
a6.set_ylabel('y')
a6.set_title(r'$y=\frac{e^x}{e^x+1}$');

fig.tight_layout()
```
