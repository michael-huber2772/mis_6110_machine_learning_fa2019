### Task 1: compute the sum of numbers from 1 to 100 using for loop (2 points)
$$S=\sum_{x=1}^{100}x$$
Hint:
1. There are many ways to solve this problem, However, you are required to use for loop.
2. You may need to set an initial value.

```python
s=0
for x in range(101):   #101 is not included
    s += x # equivalent to s = s + x
print(s)    
```

### Task 7: use list comprehension to remove stop words from a text (2 points)
`
s="This is an example that stop words can be removed from a text using the list comprehension."
s=s.lower().split()
print(s)
stopwords=['a','this','that','the','be','is','an','can','from']
`
Hint:
1. Create a list (name it as 'new_text') of words from s with stop words removed, using list comprehension.
2. join this list of words back to a string, using: 

`print(' '.join(new))`

```python
s="This is an example that stop words can be removed from a text using the list comprehension."
s=s.lower().split()
print(s)
stopwords=['a','this','that','the','be','is','an','can','from']
new_text=[x for x in s if x not in stopwords]
print(' '.join(new_text))
```

### Task 9: Create a function that calculates the sum of squares of numbers from 1 to x (3 points)
$$f(x)=\sum_{n=1}^{x} n^2$$
Hint:
1. Name your function as SQ(x).
2. Calculate $\sum_{n=1}^{10}n^2$.

```python
def SQ(x):
    """
    Calculate the sum of squares from 1 to x, x is an integer.
    """
    s=0
    for n in range(1,x+1):
        s += n**2
    return s
SQ(10)
```

### Task 10: Create a function f(x) to calculate 180/x, with considering errors (exceptions) (2 points)
$$ f(x)=\frac{180}{x} $$
Hint and requirements:
1. Name your function as 'd_180(x)'.
2. The function does not need to return a value, but has to print the result with '`print('180/'+str(x)+' =',180/x)`'.
3. Use try/except, if caught an exception, print out the exception.
3. run d_180(10), d_180(0),d_180('not a number'),d_180([1]),d_180('90')


```python
def d_180(x):
    try:
        print('180/'+str(x)+' =',180/x)
    except Exception as e:
        print("Caught an exception:" + str(e))
d_180(10)
d_180(0)
d_180('not a number')
d_180([1])
d_180('90')
```
