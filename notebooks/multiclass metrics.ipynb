# Multiclass Classification Metrics

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.metrics import precision_score, recall_score,confusion_matrix,accuracy_score,classification_report
```

```python
y_true=['a','a','a','a','b','b','b','c','c','c']
y_pred=['a','a','b','b','b','b','a','c','c','c']

#y_true=['a','a','a','a','b','b','b','c','c','c','c','c','c','c','c','c','c','c','c','c','c','c','c','c','c']
#y_pred=['a','a','b','b','b','b','a','c','c','c','c','c','c','b','b','a','b','b','a','b','b','a','b','b','a']
```

```python
confusion=confusion_matrix(y_true,y_pred)
sns.heatmap(confusion,annot=True, fmt="d",xticklabels=['a','b','c'],yticklabels=['a','b','c'])

plt.ylabel('True label')
plt.xlabel('Predicted label')
```

```python
print(classification_report(y_true,y_pred))
```

### Macro Average
+ Each class has equal weights
+ Compute metrics within each class
+ Average metrics across classes


```python
precision_score(y_true,y_pred,average='macro')
```

```python
recall_score(y_true,y_pred,average='macro')
```

### Micro Average
+ Each instance has equal weights
+ Largest classes have most influence
+ Aggregate outcomes across all classes
+ Compute metrics with aggregate outcomes


```python
precision_score(y_true,y_pred,average='micro')
```

```python
recall_score(y_true,y_pred,average='micro')
```

### Weighted Average
Mean per-class metric, weighted by support

**Therefore, it focuses on most frequent class.**

```python
precision_score(y_true,y_pred,average='weighted')
```
